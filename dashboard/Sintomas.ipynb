```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import plotly.express as px
from wordcloud import WordCloud

import base64
from io import BytesIO
```

```python
datasintomas = pd.read_csv("https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto21/SintomasCasosConfirmados.csv")
datasintomas = datasintomas[["Sintomas", "2020-07-13"]]
datasintomas.head()
```

```python
sentences = datasintomas["Sintomas"].tolist()
sentences_as_a_string = ' '.join(sentences)

# Convert the string into WordCloud
plt.figure(figsize=(10, 10))
plt.imshow(WordCloud().generate(sentences_as_a_string))
plt.axis("off")
```

```python
def plot_wordcloud (data):
        d = {}
        for a, x in data["Sintomas"].values:
            d[a] = x
        wc = WordCloud(background_color='black',
                    width=1800,
                    height=1400).generate_from_frequencies(frequencies=d)   
        return (wc)
```

```python
def get_wordcloud(data):
    sentences = data["Sintomas"].tolist()
    sentences_as_a_string = ' '.join(sentences)

    # Convert the string into WordCloud
    #plt.figure(figsize=(10, 10))
    wc = WordCloud().generate(sentences_as_a_string)
    plt.imshow(wc)
    plt.axis("off")
    wc_img = wc.to_image()
    with BytesIO() as buffer:
        wc_img.save(buffer, 'png')
        img2 = base64.b64encode(buffer.getvalue()).decode()     
    #plt.show()
    return img2
```

```python
get_wordcloud(datasintomas)
```
