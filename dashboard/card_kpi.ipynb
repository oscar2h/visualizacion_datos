```python
TotalKPI_Region = pd.read_csv("https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto4/2022-08-29-CasosConfirmados-totalRegional.csv")
TotalKPI_Region.head()
```

```python
TotalKPI_Region = TotalKPI_Region[TotalKPI_Region['Region'] == 'Arica y Parinacota']
```

```python
casos_activos = TotalKPI_Region['Casos activos confirmados'].iloc[-1]
casos_nuevos = TotalKPI_Region['Casos nuevos totales'].iloc[-1]
fallecidos = TotalKPI_Region['Fallecidos totales'].iloc[-1]
total_cum = TotalKPI_Region['Casos totales acumulados'].iloc[-1]
recuperados = TotalKPI_Region['Casos confirmados recuperados'].iloc[-1]
```

```python
def get_kpi(data,region, kpi):
    TotalKPI_Region = data[data['Region'] == region]
    TotalKPI_Region = TotalKPI_Region[kpi].iloc[-1]
    return TotalKPI_Region
```
